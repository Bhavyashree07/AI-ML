```python
import numpy as nm
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# Importing the dataset
dataset = pd.read_csv('User_Data.csv')
x = dataset.iloc[:, [2, 3]].values
y = dataset.iloc[:, 4].values
```

```python
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
type(x_train)
```

```python
print(x_train)
```

```python
# Feature Scaling
sc = StandardScaler()
x_train1 = sc.fit_transform(x_train)
x_test1= sc.transform(x_test)
```

```python
mn=MinMaxScaler()
x_train2 = mn.fit_transform(x_train)
x_test2 = mn.transform(x_test)
```

```python
print(x_train1)
```

```python
print(x_train2)
```

```python
xtrain=x_train1+x_train2
print(xtrain)
```

```python
x_train1.shape
```

```python
feature_train=pd.concat([pd.DataFrame(x_train1),pd.DataFrame(x_train2)],axis=1)
```

```python
feature_test=pd.concat([pd.DataFrame(x_test1),pd.DataFrame(x_test2)],axis=1)
```

```python
# Fitting I Bayes to the Training set
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
# Fitting I Bayes to the Training set
classifier1 = GaussianNB()
classifier1.fit(feature_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)
```

```python
# Predicting the Test set results
y_pred1 = classifier1.predict(feature_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix : \n",cm)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred1)
print("Confusion Matrix : \n",cm)
```

```python
print ("Accuracy : ", accuracy_score(y_test, y_pred))  
```

```python
print ("Accuracy : ", accuracy_score(y_test, y_pred1))  
```

```python
from sklearn import metrics
import numpy as np
```

```python
y_pred=np.around(classifier.predict(x_test))
print(metrics.classification_report(y_test,y_pred))
```
